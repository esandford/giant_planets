```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')

import baraffe_tables
from baraffe_tables.table_search import baraffe_table_search

from astropy.table import Table

%matplotlib inline
```

```python
# read in MESA history file for stalled run
hist = Table.read("./Minit=1.6Mj/LOGS/history.data", format="ascii", header_start=4, data_start=5)
print(hist.colnames)
print(len(hist.colnames))
```

```python
for i in range(1, len(hist.colnames)):

    fig, ax = plt.subplots(1,1,figsize=(8,6))
    ax.plot(hist['model_number'],hist[hist.colnames[i]], 'k.')
    ax.set_xlabel('model number')
    ax.set_ylabel('{0}'.format(hist.colnames[i]))
    #ax.set_xlim(80,250)
    #ax.axvline(226,color='r')
    #ax.axvline(229,color='r')
    #ax.axvline(232,color='g')
    if hist.colnames[i] == 'Teff':
        ax.axhline(100, color='b')
    plt.show()
    plt.close()
```

```python
for i in range(1, len(hist.colnames)):

    fig, ax = plt.subplots(1,1,figsize=(8,6))
    ax.plot(hist['star_age'],hist[hist.colnames[i]], 'k.')
    ax.set_xlabel('star age')
    ax.set_ylabel('{0}'.format(hist.colnames[i]))
    #ax.set_xlim(80,250)
    #ax.axvline(226,color='r')
    #ax.axvline(229,color='r')
    #ax.axvline(232,color='g')
    ax.set_xscale("log")
    if hist.colnames[i] == 'Teff':
        ax.axhline(100, color='b')
    plt.show()
    plt.close()
```
