```python
#jupiter masses
Minit = np.round(np.logspace(np.log10(0.2),np.log10(20.), num=20),2)[::-1]
print(Minit)
print(len(Minit))
#jupiter radii
Rinit = np.array((1.75,1.5,1.25))
```

```python
os.chdir('/Users/emily/Documents/astro/giant_planets/science_runs/1_created_models/')

for i, m in enumerate(Minit):
    for j, r in enumerate(Rinit):
        print("Minit is {0}, Rinit is {1}".format(m,r))
        
        inlist_create = mh.Inlist('./inlist_create')
        create_model_filename = "planet_create_{0}_Mj_{1}_Rj.mod".format(m, r)
        create_profile_filename = "planet_create_{0}_Mj_{1}_Rj.profile".format(m, r)

        inlist_create.set_option('mass_in_gm_for_create_initial_model', m*mJtoGrams)
        inlist_create.set_option('radius_in_cm_for_create_initial_model', r*rJtoCm)
        inlist_create.set_option('save_model_filename', create_model_filename)
        inlist_create.set_option('filename_for_profile_when_terminate', create_profile_filename)

        # edit rn script
        f = open('./rn_template'.format(m), 'r')
        g = f.read()
        f.close()
        #print(g)
        g = g.replace("<<create_model_filename>>", create_model_filename)
        
        h = open('./rn_temp'.format(m), 'w')
        h.write(g)
        h.close()
        shutil.copyfile('./rn_temp'.format(m),'./rn'.format(m))
        
        # execute mk, rn scripts
        os.system('./mk')
        os.system('./rn')

        print("")
        print("")
```
